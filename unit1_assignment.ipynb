```python
from transformers import pipeline
```

```python
models = {
    "bert": "bert-base-uncased",
    "roberta": "roberta-base",
    "bart": "facebook/bart-base"
}
```

Experiment 1: Text Generation

```python
prompt = "The future of Artificial Intelligence is"

for name, model in models.items():
    print(f"\nModel: {name}")
    try:
        generator = pipeline("text-generation", model=model)
        output = generator(prompt, max_length=30)
        print(output)
    except Exception as e:
        print("Error:", e)
```

Experiment 2: Masked Language Modelling (Fill-Mask)

```python
sent_bert = "The goal of Generative AI is to [MASK] new content."
sent_bart = "The goal of Generative AI is to <mask> new content."

# BERT
fill_bert = pipeline("fill-mask", model=models["bert"])
print("BERT:", fill_bert(sent_bert))

# RoBERTa
fill_roberta = pipeline("fill-mask", model=models["roberta"])
print("\nRoBERTa:", fill_roberta(sent_bert))

# BART
fill_bart = pipeline("fill-mask", model=models["bart"])
print("\nBART:", fill_bart(sent_bart))
```

Experiment 3: Question Answering 

```python
context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
question = "What are the risks?"

for name, model in models.items():
    print(f"\nModel: {name}")
    try:
        qa = pipeline("question-answering", model=model)
        result = qa(question=question, context=context)
        print(result)
    except Exception as e:
        print("Error:", e)
```

Observation Table
| Task        | Model     | Classification (Success/Failure) | Observation (What actually happened?) | Why did this happen? (Architectural Reason) |
|------------|-----------|----------------------------------|----------------------------------------|---------------------------------------------|
| Generation | BERT      | Failure | The text-generation pipeline fails or throws an error and cannot generate new text. | BERT is an encoder-only model and is not designed for autoregressive next-token prediction. |
| Generation | RoBERTa   | Failure | Similar to BERT, the model fails to generate text and the pipeline errors out. | RoBERTa is also an encoder-only architecture without a decoder for text generation. |
| Generation | BART      | Success | Generates a coherent continuation of the prompt, though the output is generic. | BART is an encoder–decoder model trained for sequence-to-sequence and generative tasks. |
| Fill-Mask  | BERT      | Success | Correctly predicts words such as “create” or “generate” for the masked token. | BERT is trained using Masked Language Modeling (MLM), making it well-suited for this task. |
| Fill-Mask  | RoBERTa   | Success | Produces highly confident and accurate mask predictions. | RoBERTa improves upon BERT’s MLM training with more data and optimized training strategies. |
| Fill-Mask  | BART      | Partial Success | Can fill the masked token, but predictions are less precise than BERT/RoBERTa. | BART is trained with a denoising objective rather than pure MLM, so this is not its primary strength. |
| QA         | BERT      | Partial Success | Returns partial or incomplete answers such as a single risk term. | The base BERT model is not fine-tuned for extractive Question Answering tasks like SQuAD. |
| QA         | RoBERTa   | Partial Success | Produces similar partial answers, sometimes selecting a better text span. | Strong encoder representations help, but lack of QA fine-tuning limits performance. |
| QA         | BART      | Partial Success | Answers are vague or inconsistent and may not directly match the expected span. | BART is optimized for text generation rather than extractive Question Answering. |
